### 처음부터 배열 만들기
규모가 큰 배열의 경우에 numpy에 내장된 루틴을 사용해 처음부터 배열을 생성하는 것이 더 효율적임

```python
# 0으로 채워진 10의 정수 배열 만들기 
np.zeros(10,dtype=int)

# 1로 채워진 3*5 부동소수점 배열 만들기 
np.full((3,5), 1, dtype=float)

# 선형 수열로 채운 배열 만들기(0~20까지 2식 더한 배열)
np.arange(2,20,2)

# 0과 2 사이에 일정한 간격을 가진 다섯 개의 값으로 채운 배열 만들기 
np.linspace(0,2,5)

# 균등하게 분포된 3*3 배열 만들기(0~1(고정) 난수)
np.random.random((3,3))

# 정규분포 m=2, sd = 0.5 의 난수로 채운 3*3배열
np.random.normal(2,0.5,(3,3))

# 1~20 구간의 임의의 정수로 채운 4*4 배열 만들기 
np.random.randint(1,20,(4,4))

# 2*2 unit matrix
np.eye(2)

# 다섯 정수를 가지는 초기화되지 않은 배열 만들기
# 같은 해당 메모리 위치에 이미 존재하고 있는 값으로 채움
np.empty(5)
```

### numpy 배열의 기초
배열 속성 지정: 배열의 크기, 모양, 메모리 소비량, 데이터 타입을 결정한다.<br>
배열 인덱싱: 개별 배열 요솟값을 가져오고 설정 <br>
배열 슬라이싱: 큰배열 내에 있는 작은 하위 배열을 가져오고 설정 <br>
배열 재구조화: 해당 배열의 형상을 변경한다. <br>
배열 결합 및 분할: 여러 배열을 하나로 결합하고 하나의 배열을 여러 개로 분할한다.

```python
# 배열 속성 지정 
np.random.seed(28) # 재현을 위한 시드값 지정
x1 = np.random.randint(10,size=6) # 1-dim
# x2 = np.random.randint(10,size=(3,3)) # 2-dim
# x3 = np.random.randint(10,size=(3,4,5)) # 3-dim
# x3
# x2
x1
```

```python
# 배열 인덱싱: 단일 요소 접근
x1[0] # 첫 번째 값 
x1[-1] # 끝에서 인덱싱 하려면 음수값을 넣어야함

# 다차원 배열에서 콤마로 구분된 인데스 튜플을 이용해 접근 가능함
# print('{}'.format(x2)) 
# print('{}'.format(x2[1,0]))

# 인덱스 표기법을 사용해 값을 수정 가능 
x2[2,1] = 10 # 소수점 표기시 소수점 이하는 잘림
x2
```

```python
# 배열 슬라이싱: 하위 배열에 접근하기 
# x[start:stop:step]
# 단일 차원 
x1[:3] # 세 번째 요소까지 
# 다차원 
# x2[:2, :3] # 두 개의 행, 세 개의 열
# 다차원 역 변환 
# x2[::-1, ::-1]

# 사본이 아닌 뷰로서 하위 배열 
# 배열 슬라이스는 배열 데이터의 뷰를 반환함 
print(x2)
x2_sub = x2[:2,:2] # 2*2 하위 배열
print(x2_sub)
```

```python
# 배열 사본 만들기 
x2_copy = x2[:2,:2].copy()
print(x2_copy)
```

```python
# 배열 재구조화 
g = np.arange(1,10).reshape((3,3))
g
# 연속되지 않은 메모리 버퍼일 경우 그렇지 않을 수 있음
g[:, np.newaxis] # 열벡터로 재 구조화
```

```python
# 배열 연결 및 분할 
# 배열 연결 
x = np.array([1,2,3])
y = np.array([3,2,1])
np.concatenate([x,y]) # 연결

td = np.array([[4,3,2],
              [8,6,9]])
np.concatenate([td,td],axis=1) # 두 번째 축을 따라 연결 

# 수직으로 쌓을 때 
np.vstack([x,td])
# 수평으로 쌀음
z = np.array([[94],
             [30]])
np.hstack([td,z])
```
